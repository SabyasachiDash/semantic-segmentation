# Exploratory Data Analysis

#### KITTI Dataset

http://www.cvlibs.net/datasets/kitti/eval_road.php

From their website:

The road and lane estimation benchmark consists of 289 training and 290 test images. It contains three different categories of road scenes:

- uu - urban unmarked (98/100)
- um - urban marked (95/96)
- umm - urban multiple marked lanes (96/94)
- urban - combination of the three above

Ground truth has been generated by manual annotation of the images and is available for two different road terrain types: 

- road - the road area, i.e, the composition of all lanes, and 
- lane - the ego-lane, i.e., the lane the vehicle is currently driving on (only available for category "um"). 

Ground truth is provided for training images only.

##### Reference Paper

- http://www.cvlibs.net/publications/Fritsch2013ITSC.pdf


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.style.use('ggplot')
```

```python
def show_image(img, title):
    plt.imshow(img)
    plt.axis('off')
    plt.title(title)
    plt.show()
```

### Urban Unmarked

```python
uu_img = cv2.imread('data/data_road/training/image_2/uu_000000.png')
uu_img = cv2.cvtColor(uu_img, cv2.COLOR_BGR2RGB)

uu_road_img = cv2.imread('data/data_road/training/gt_image_2/uu_road_000000.png')
uu_road_img = cv2.cvtColor(uu_road_img, cv2.COLOR_BGR2RGB)

uu_road_overlay = cv2.addWeighted(src1=uu_img, alpha=1.0,
                                   src2=uu_road_img, beta=0.4, gamma=1)
```

```python
show_image(uu_img, 'Urban Unmarked')
show_image(uu_road_img, 'Road Label')
show_image(uu_road_overlay, 'Road Overlay')
```

### Urban Marked

```python
um_img = cv2.imread('data/data_road/training/image_2/um_000000.png')
um_img = cv2.cvtColor(um_img, cv2.COLOR_BGR2RGB)

um_lane_img = cv2.imread('data/data_road/training/gt_image_2/um_lane_000000.png')
um_lane_img = cv2.cvtColor(um_lane_img, cv2.COLOR_BGR2RGB)

um_road_img = cv2.imread('data/data_road/training/gt_image_2/um_road_000000.png')
um_road_img = cv2.cvtColor(um_road_img, cv2.COLOR_BGR2RGB)

um_lane_overlay = cv2.addWeighted(src1=um_img, alpha=1.0,
                                  src2=um_lane_img, beta=0.4, gamma=1)

um_road_overlay = cv2.addWeighted(src1=um_img, alpha=1.0,
                                  src2=um_road_img, beta=0.4, gamma=1)
```

```python
show_image(um_img, 'Urban Marked Image')

show_image(um_lane_img, 'Lane Label')
show_image(um_lane_overlay, 'Lane Overlay')

show_image(um_road_img, 'Road Label')
show_image(um_road_overlay, 'Road Overlay')
```

### Urban Multiple Marked

```python
umm_img = cv2.imread('data/data_road/training/image_2/umm_000000.png')
umm_img = cv2.cvtColor(umm_img, cv2.COLOR_BGR2RGB)

umm_road_img = cv2.imread('data/data_road/training/gt_image_2/umm_road_000000.png')
umm_road_img = cv2.cvtColor(umm_road_img, cv2.COLOR_BGR2RGB)

umm_road_overlay = cv2.addWeighted(src1=umm_img, alpha=1.0,
                                   src2=umm_road_img, beta=0.4, gamma=1)
```

```python
show_image(umm_img, 'Urban Multiple Marked')
show_image(umm_road_img, 'Road Label')
show_image(umm_road_overlay, 'Road Overlay')
```
